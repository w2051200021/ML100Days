## 請搜尋目標為多標籤問題 (Multi-label) 的機器學習案例，了解其資料來源、目標以及評估指標為何
## Hint: 服飾

#### [參考來源](https://towardsdatascience.com/journey-to-the-center-of-multi-label-classification-384c40229bff)

#### [服飾影像資料集](https://www.kaggle.com/trolukovich/apparel-images-dataset)
> 資料包含不同顏色、不同樣式的服飾（例如：黑/藍/棕/紅/白 色的衣服、洋裝、褲、鞋）

#### 將資料集的標籤分成：顏色 與 服飾，例如：黑色洋裝（black_dress)，藍色衣服（blue_short）
> 進行多標籤分類（Multi-label classification)。[參考資料](https://www.kaggle.com/trolukovich/multi-label-classification-keras)

> 評估指標：分類準確率（Accuracy）
