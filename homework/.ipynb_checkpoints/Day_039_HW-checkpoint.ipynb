#### 1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection
> 要了解為什麼 LASSO Regression 可以做 Feature selection，我們可以從他最佳化的公式來看：

\begin{equation*}
\text{LASSO  Regression :} \quad \frac{1}{2}\sum_{i=1}^N(y_i-\beta_0-\sum_{j=1}^px_ij\beta_j)^2+\lambda\sum_{j=1}^p|\beta_j|
\end{equation*}

> 相對 Rigde 使用二次方的係數懲罰項，LASSO 使用係數的絕對值來做為懲罰項。直觀來說，若要找到最小的係數使上式總和最小，要達到同樣的誤差值，做 Ridge 得出來的參數一定會比 LASSO 大（因為平方項使小於零的值更快變小）。而數學上也可以證明，Ridge 只能將係數縮減到接近零但不為零，LASSO 則可以將係數降至零。因此，LASSO 可以被用來作為 Feature selection 的工具！

#### 2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?
> 可以，當變數間存在高度共線性時（變數之間可以表示成線性組合），會使係數值變得異常的大，讓模型對於輸入值 x 變得十分敏感。使用 Ridge Regression 可以設定懲罰參數來限制係式增長，使模型平滑一點，減少共線性造成的不穩定。

\begin{equation*}
w=(X^TX+\alpha I)^{-1}X^Ty
\end{equation*}

> * 將原誤差函數加入係數平方的懲罰項後，微分求解得到上式。可以看到，因為懲罰項作用在矩陣的對角線上，如果 alpha 越大，會使反矩陣的值越小，進而造成係數的縮減。
