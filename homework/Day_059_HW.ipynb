### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
plt.style.use('seaborn')

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
print(f'shape of X: {X.shape}')
print(f'shape of y: {y.shape}')
```

畫出將原始資料

```python
numbers_to_display = 25

# Calculate the number of cells that will hold all the numbers.
num_cells = 5

# Make the plot a little bit bigger than default one.
plt.figure(figsize = (10, 10))
# Go through the first numbers in a training set and plot them.
for plot_index in range(numbers_to_display):
    # Extrace digit data.
    digit = X[plot_index, :]
    digit_label = y[plot_index]

    # Calculate image size (remember that each picture has square proportions).
    image_size = int(math.sqrt(digit.shape[0]))
    
    # Convert image vector into the matrix of pixels.
    frame = digit.reshape((image_size, image_size))
    
    # Plot the number matrix.
    plt.subplot(num_cells, num_cells, plot_index + 1)
    plt.imshow(frame, cmap='Greys')
    plt.title(digit_label)
    plt.tick_params(axis='both', which='both', bottom=False, left=False, labelbottom=False, labelleft=False)

# Plot all subplots.
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

pca

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
print(f'shape of X_pca: {X_pca.shape}')
```

```python
digit_type = np.unique(y)
plt.figure(figsize = (10, 8))
for digit in digit_type:
    plt.scatter(X_pca[y == digit, 0], X_pca[y == digit, 1], label = digit)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.title('Digit data after Principal Component Analysis')
plt.show()
```

經過 PCA 降維度後，以前兩名主成分畫散佈圖可以看到三種數字各有明顯的分佈，代表 PCA 取出主成分對於資料分析是有幫助的。然而仔細分析後不難發現，之所以這筆資料在降維度之後能夠明顯分開，是因為每一張數字圖片都是用同樣大小的紙，在差不多的位置寫下，猜想若今天是用一張很大的紙，每次都寫在不同位置，以這樣的分法可能就無法判別。
