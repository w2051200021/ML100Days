# 作業

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

參考 sklearn 的官方 document [Kmeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans)

```python
plt.style.available # 可用的各種繪圖樣式
```

### 測試不同的群數

```python

randomState = 42
nrows = 2
ncols = 3

plt.figure(figsize = (18, 12))

# 嘗試各種不同的群數
for i in range(1, 7):
    kmeans = KMeans(n_clusters = i, random_state = randomState)
    y_pred = kmeans.fit_predict(X)
    
    plt.subplot(nrows, ncols, i)
    plt.scatter(X[:, 0], X[:, 1], c = y_pred)
    
    plt.title(f'{i}-means cluster')
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

### 嘗試不同的 init (初始條件)

```python
randomState = 42
nrows = 1
ncols = 2

plt.figure(figsize = (12, 6))

# 嘗試各種不同的 
# init: kmeans++ (default)

kmeans = KMeans(n_clusters = 3, random_state = randomState, init = 'k-means++')
y_pred = kmeans.fit_predict(X)

plt.subplot(nrows, ncols, 1)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)

plt.title(f'3 - means cluster (k-means++)')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# init: random
kmeans = KMeans(n_clusters = 3, random_state = randomState, init = 'random')
y_pred = kmeans.fit_predict(X)

plt.subplot(nrows, ncols, 2)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)

plt.title(f'3 - means cluster (random)')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

可能因為資料比較簡單，所以使用兩種不同的初始條件都得到一樣的結果。
