```python
# 做完特徵工程前的所有準備
import pandas as pd
import numpy as np
import copy
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
# 因為擬合(fit)與編碼(transform)需要分開, 因此不使用.get_dummy, 而採用 sklearn 的 OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

data_path = 'C:/Users/user/Desktop/ML marathon/data/Feature Engineering/'
df = pd.read_csv(data_path + 'titanic_train.csv')

train_Y = df['Survived']
df = df.drop(['PassengerId', 'Survived'] , axis=1)
df.head()
```

```python
# 因為需要把類別型與數值型特徵都加入, 故使用最簡版的特徵工程
LEncoder = LabelEncoder()
MMEncoder = MinMaxScaler()
for c in df.columns:
    df[c] = df[c].fillna(-1)
    if df[c].dtype == 'object':
        df[c] = LEncoder.fit_transform(list(df[c].values))
    df[c] = MMEncoder.fit_transform(df[c].values.reshape(-1, 1))
df.head()
```

```python
train_X = df.values
# 因為訓練邏輯斯迴歸時也要資料, 因此將訓練及切成三部分 train / val / test, 採用 test 驗證而非 k-fold 交叉驗證
# train 用來訓練梯度提升樹, val 用來訓練邏輯斯迴歸, test 驗證效果
train_X, test_X, train_Y, test_Y = train_test_split(train_X, train_Y, test_size=0.5)
train_X, val_X, train_Y, val_Y = train_test_split(train_X, train_Y, test_size=0.5)
```

# 作業1
* 請對照範例，完成隨機森林的鐵達尼生存率預測，以及對應的葉編碼+邏輯斯迴歸

```python
# 隨機森林擬合後, 再將葉編碼 (*.apply) 結果做獨熱 / 邏輯斯迴歸
rf = RandomForestClassifier(n_estimators=20, min_samples_split=10, min_samples_leaf=5, 
                            max_features=4, max_depth=3, bootstrap=True, random_state = 42)
onehot = OneHotEncoder()
lr = LogisticRegression(solver='lbfgs', max_iter=1000)

"""
Your Code Here (Hint : 隨機森林的葉編碼(.apply)不需要加上[:, :, 0], 直接用rf.apply()調用即可, 本作業其餘寫法相同)
"""
rf.fit(train_X, train_Y) # 訓練隨機森林的模型
onehot.fit(rf.apply(train_X)) # 將訓練過的隨機森林特徵做獨熱編碼
lr.fit(onehot.transform(rf.apply(val_X)), val_Y) # 再用另一筆訓練資料代入隨機森林模型得到分類特徵，
                                                 # 以獨熱編碼轉換後，用以訓練邏輯迴歸模型
```

#### [ROC_curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html#sklearn.metrics.roc_curve) : Make it easy to identify the best threshold
> 輸入真實值與預測值，調整不同的閾值，觀察 True Positive Rate 和 False Positive Rate 之間的關係。
* 閾值的意義：分類模型會給出該輸入值符合某特定分類的機率，閾值的大小會影響分類，例如：閾值 = 0（或者預測值中最小者），所有輸入值都會被判定為陽性，因此真陽性和假陽性的機率都會是 1，位於 ROC 圖的右上角。

```python
# 將隨機森林+葉編碼+邏輯斯迴歸結果輸出
pred_rf_lr = lr.predict_proba(onehot.transform(rf.apply(test_X)))[:, 1] # 只取生存下來的機率 :1
fpr_rf_lr, tpr_rf_lr, _ = roc_curve(test_Y, pred_rf_lr) # 通常以 _ 存取不用到的東西，在這邊是參數的閾值－從大到小排列 

# 將隨機森林結果輸出
pred_rf = rf.predict_proba(test_X)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(test_Y, pred_rf)
```

# 作業2
* 上述的結果，葉編碼是否有提高預測的正確性呢?
> 以隨機森林的結果來看，葉編碼的效果並沒有顯著的好。

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.figure(figsize=(12,12))
# 將結果繪圖
plt.plot([1, 0], [1, 0], 'k--')
plt.plot(fpr_rf_lr, tpr_rf_lr, label = 'RF + LR')
plt.plot(fpr_rf, tpr_rf, label = 'RF')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```

#### [Area Under Curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_auc_score.html#sklearn.metrics.roc_auc_score)
> 若以 AUC 當作判斷標準，亦顯示葉編碼對於此分類問題效果沒有比較好。

```python
# RF + LR
roc_auc_score(test_Y, pred_rf_lr)
```

```python
# RF
roc_auc_score(test_Y, pred_rf)
```
