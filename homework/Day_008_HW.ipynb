#### 確認所有欄位資料型態

#### 分出不同資料型態的欄位

```python
int_features = []
float_features = []
object_features = []

for feature, dtype in zip(app_train.columns, app_train.dtypes):
    if dtype == 'int64':
        int_features.append(feature)
    elif dtype == 'float64':
        float_features.append(feature)
    else:
        object_features.append(feature)
```

```python
app_train[int_features].head()
```

```python
app_train[float_features].head()
```

```python
app_train[object_features].head()
```
